```python
import pandas as pd
import numpy as np
df = pd.read_csv('olympics-1986-2016-cleaned.csv')
df.head()
```

```python
# THINGS TO DO:
# Make sure that there only exists 1 row for each OLYMPIAN
# Group by SPORT -> Aggregate (mean) by AGE
```

```python
# Make sure that there only exists 1 row for each OLYMPIAN
df = df.drop_duplicates(subset='Name', keep='first')
```

```python
# Group by SPORT -> Aggregate (mean) by AGE
x = df.groupby(['Sport', 'Sex'])['Age'].mean()
df_new = pd.DataFrame(x)
df_new = df_new.reset_index()
df_new.Age = df_new.Age.round()
df_new.head()
```

```python
df_M = df_new.loc[df['Sex'] == "M"]
df_F = df_new.loc[df['Sex'] == "F"]

# sorted(list(set(df_new.Sport.unique()) - set(df_M.Sport.unique())))
# sorted(list(set(df_new.Sport.unique()) - set(df_F.Sport.unique())))
```

```python
for sport in sorted(list(set(df_new.Sport.unique()) - set(df_M.Sport.unique()))):
    df_new = df_new.append({'Sport':sport, 'Sex':'M', 'Age':0}, ignore_index=True)

for sport in sorted(list(set(df_new.Sport.unique()) - set(df_F.Sport.unique()))):
    df_new = df_new.append({'Sport':sport, 'Sex':'F', 'Age':0}, ignore_index=True)
```

```python
df_new.to_csv("olympics-1986-2016-cleaned-3.csv", sep=',', index=False)
```
