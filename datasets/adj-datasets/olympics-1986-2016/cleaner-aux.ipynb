```python
def list_to_dict(lst):
    res_dct = {float(lst[i]): lst[i + 1] for i in range(0, len(lst), 2)}
    return res_dct

def sorter(ocean_score):
    ocean_score = ocean_score.split()
    del ocean_score[1::3]
    ocean_score.reverse()
    return list_to_dict(ocean_score)
```
