```python
import pandas as pd
df_main = pd.read_csv('olympics-1986-2016.csv')
df_main.head()
```

```python
df_noc = pd.read_csv('NOC-countries-abbreviations.csv')
df_noc.head()
```

```python
# Join df_noc to df using "NOC" as key
df = pd.merge(df_main, df_noc, on='NOC', how='left')
df = df[df['Country'].notna()]
df
```

```python
print(df_main.shape)
print(df_noc.shape)
print(df.shape)
```

```python
df['NOC'].value_counts()
```

```python
df.to_csv("olympics-1986-2016-cleaned.csv", sep=',', index=False)
```
