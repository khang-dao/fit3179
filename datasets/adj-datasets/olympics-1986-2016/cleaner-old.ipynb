```python
import pandas as pd
df = pd.read_csv('olympics-1986-2016.csv')
df.head(3)
```

```python
df1 = (df.groupby(['Team'])
         .apply(lambda x: (pd.notna(x['Medal'])).sum())
         .reset_index(name='count')
         )
df2 = df1.sort_values(by='count', ascending=False).reset_index(drop=True)
df2.head(10)
```

```python
df2.to_csv("olympics-1986-2016-cleaned.csv", sep=',', index=False)
```
