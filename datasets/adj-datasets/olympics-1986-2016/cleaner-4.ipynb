```python
import pandas as pd
df = pd.read_csv('olympics-1986-2016-cleaned.csv')
df.replace({'Chinese Taipei': 'Taiwan', 'Hong Kong, China': 'Hong Kong'}, inplace=True)
df.head(5)
```

```python
# THINGS TO DO:
# Make sure that there only exists 1 row for each OLYMPIAN
# Delete rows with NA values (HEIGHT/WEIGHT)
# Assign a CONTINENT to each row
```

```python
# Make sure that there only exists 1 row for each OLYMPIAN
df = df.drop_duplicates(subset='Name', keep='first')
df = df[df['Height'].notna()]
df = df[df['Weight'].notna()]
df = df[['Country', 'Height','Weight']]
```

```python
from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
continents = {'NA': 'America', 'SA': 'America', 'AS': 'Asia', 'OC': 'Oceania', 'AF': 'Africa', 'EU': 'Europe'}
countries = sorted(df['Country'].unique())
```

```python
bad_countries = []
countries_with_continents = {}

for country in countries:
    try:
        # countries_with_continents.append((country, continents[country_alpha2_to_continent_code(country_name_to_country_alpha2(country))]))
        countries_with_continents[country] =  continents[country_alpha2_to_continent_code(country_name_to_country_alpha2(country))]
    except:
        bad_countries.append(country)
        countries.remove(country)
        print(country)
```

```python
countries_with_continents['Kosovo'] = 'Europe'
countries_with_continents['The Gambia'] = 'Africa'
countries_with_continents['Virgin Islands'] = 'America'
countries_with_continents['Timor-Leste'] = 'Asia'
countries_with_continents['Kuwait'] = 'Asia'
countries_with_continents['Yemen'] = 'Asia'
```

```python
def get_continent(country):
    return countries_with_continents[country]
df['Continent'] = df['Country'].apply(lambda x: get_continent(x))
df = df[['Continent', 'Height','Weight']]
df.head()
```

```python
df_asia = df[df['Continent'] == 'Asia']
df_america = df[df['Continent'] == 'America']
df_africa = df[df['Continent'] == 'Africa']
df_europe = df[df['Continent'] == 'Europe']
df_oceania = df[df['Continent'] == 'Oceania']
```

```python
df_asia.to_csv("olympics-1986-2016-cleaned-4-asia.csv", sep=',', index=False)
df_america.to_csv("olympics-1986-2016-cleaned-4-america.csv", sep=',', index=False)
df_africa.to_csv("olympics-1986-2016-cleaned-4-africa.csv", sep=',', index=False)
df_europe.to_csv("olympics-1986-2016-cleaned-4-europe.csv", sep=',', index=False)
df_oceania.to_csv("olympics-1986-2016-cleaned-4-oceania.csv", sep=',', index=False)
```
